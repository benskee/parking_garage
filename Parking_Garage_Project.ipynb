```python
from IPython.display import clear_output

class Spaces: 
    empty_spaces = list(range(1,6))
    
class Ticket:
    tickets = list(range(1,6))
    
class Customer:
          
    def __init__(self):

        self.current_ticket = {'paid' : False}
        self.dashboard = []
        self.parking_spot = []
        
    def take_ticket(self, ticketer):
        self.dashboard.append(ticketer.tickets.pop())
        
    def take_space(self, spacer):
        self.parking_spot.append(spacer.empty_spaces.pop())
    
    def return_ticket(self, ticketer, ticket_number):
        ticketer.tickets.append(ticket_number)
        self.dashboard.remove(ticket_number)
      
    def return_space(self, spacer, ticket_number):
        spacer.empty_spaces.append(ticket_number)
        self.parking_spot.remove(ticket_number)
        
    
class Park:
    def parking(self):
        customer = Customer()
        spaces = Spaces()
        ticket = Ticket()
        quit = False
        marks = []
        i = 0
        paid_tickets = []
            
        while  quit == False:
            
            print("Welcome to Ben's Parktastic Parking Lot. Where parking is a whole LOT of fun!!")
            
            while True:
                choice = input('Would you like to pay or park?(enter q to quit)\n')
                if choice.lower() == 'q':
                    quit = True
                    break

                elif choice.lower() == 'park':
                    if len(ticket.tickets) == 0:
                        print('Sorry we are full! Please come back later.\n')
                        break
                    
                    else:
                        customer.take_ticket(ticket)
                        customer.take_space(spaces)
                        marks.append(customer)
                        print('Here is your ticket.\n')
                        print(f'Your ticket number is: {customer.dashboard[-1]}')
                        print(f'Your parking space number is: {customer.parking_spot[-1]}\n\n')
                        input('Press any key to continue.')
                        clear_output()
                        break


                elif choice.lower() == 'pay':
                    ticket_number = int(input('\nWhat is your ticket number? '))
                    x = list(range(1,6))                    
                    if ticket_number in x:
                            if ticket_number not in ticket.tickets:
                                
                                while True:
                                    payment = input("\nThat will be 50,000 Sats please. Press y to accept the transaction. ")
                                   
                                    if payment.lower() == 'y':
                                        print('\nThank you! Your ticket is paid. You have 15 minutes to leave. Enjoy your day!\n')
                                        paying_customer = marks.pop()
                                        paying_customer.return_ticket(ticket, ticket_number)
                                        paying_customer.return_space(spaces, ticket_number)
                                        paying_customer.current_ticket['paid'] = True
                                        paid_tickets.append(paying_customer)
                                        input('Press any key to continue.')
                                        clear_output()
                                        print("Welcome to Ben's Parktastic Parking Lot. Where parking is a whole LOT of fun!!")
                                        break
                                    else:
                                        print("You must pay to exit. Please try again.\n")
                                        
                            else:       
                                print("Sorry I don't have a record for that ticket.\n")
                                break
                                
                    else:       
                        print("Sorry I don't have a record for that ticket.\n")
                        break
                else:
                    print("Please try again.\n")


        
            
park = Park()
park.parking()
    
    
    
    
    
```
